```python
df=pd.read_csv('/content/Population.csv')
df
```

```python
X=np.array(df['year']).reshape(-1,1)
#we need a column vector but np.array will convert it into a array
#reshape will convert row vector to column vector
X
```

```python
y=np.array(df['population']).reshape(-1,1)
y
```

```python
#we'll have to send 80% data for training and remaining 20% for testing
# we need to perform train_test_split
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,y,test_size=.2)
# year is along X-axis and population is along Y-axis
```

```python
Xtrain#random selection
```

```python
Xtest#for testing the output purposes
```

```python
Ytrain#populations for training purposes cooresponding to the X-axis year values
```

```python
#we will be using linear regression
from sklearn.linear_model import LinearRegression
lg=LinearRegression()
lg.fit(Xtrain,Ytrain)
```

```python
lg.intercept_ #prints the y-intercept value (c)
```

```python
lg.coef_ #prints the value of slope(m) in y=mx+c
```

```python
ypred=lg.predict(Xtest)
print(Ytest,ypred)
```

```python
from sklearn import metrics
```

```python
print("Mean absolute Error= ",metrics.mean_absolute_error(Ytest,ypred))
```

```python
print("the mean square error is= ",metrics.mean_squared_error(Ytest,ypred))
```

```python
print("Root mean squared error is= ",np.sqrt(metrics.mean_squared_error(Ytest,ypred)))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(Xtest,Ytest,color='r')
plt.scatter(Xtest,ypred,color='g')
plt.title(label="population prediction")
```
