```python
import pandas as pd
import numpy as np
df=pd.read_csv('/content/wine quality.csv')
df.head(5)
```

```python
df.shape#number of rows and columns
```

```python
df.describe()
```

```python
df.isnull().any()#to check whether there is any null values present or not
```

```python
x=df[['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','sulphates','alcohol']].values
x
```

```python
x[0:5]
```

```python
y=df['quality'].values
y[0:5]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2)
```

```python
from sklearn.linear_model import LinearRegression
lg=LinearRegression()
lg.fit(xtrain,ytrain)
```

```python
ypred=lg.predict(xtest)
df=pd.DataFrame({'Actual':ytest,'Predicted':ypred})
df
```

```python
from sklearn import metrics
print("Mean square error",metrics.mean_squared_error(ytest,ypred))
```
