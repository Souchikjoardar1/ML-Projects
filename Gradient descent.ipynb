```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#lambda notations in python
function=lambda X:(x+5)**2#to represent a function in python
```

```python
#linespace states the values from which x starts here -10 to 10 with the precision of a 1000 points
x=np.linspace(-10,0,1000)
plt.plot(x,function(x)) #plots x and y here y being f(x)
plt.show()
```

```python
#initial value of x (x0)
x0=3#random value
lr=0.01#learning rate
itr=1 #interating variable
```

```python
df=lambda x:2*(x+5)#function for derivation of (x+5)**2
```

```python
#iterate upto 400 times it will tend to reach -5 but it won't hit the exact value
while itr<=400:
  x1=x0-lr*df(x0)
  print("iteration ",itr,"x1 ",x1)
  x0=x1
  itr=itr+1
```
